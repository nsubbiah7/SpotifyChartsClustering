```python
from sklearn.cluster import KMeans

# Group countries into clusters based on K-means clustering

def kmeans_clustering(data, num_clusters):
    # Initialize KMeans object
    kmeans = KMeans(n_clusters=num_clusters, n_init=10)
    # Fit the KMeans model to the data
    kmeans.fit(data)
    # Get cluster labels for each data point
    labels = kmeans.labels_
    # Get centroids of clusters
    centroids = kmeans.cluster_centers_
    return labels, centroids

data = [argentina, australia, austria, belarus, belgium, bolivia, brazil, bulgaria, canada, chile]
labels, centroids = kmeans_clustering(data, 5)
print("Cluster Labels:", labels)
print("Centroids:", centroids)
```

```python
feature_names = ['danceability', 'energy', 'loudness', 'acousticness', 'instrumentalness', 'liveness', 'valence']
def cluster_analysis(data, labels, feature_names):
    # Initialize dictionary to store attributes of clusters
    cluster_attributes = {i: [] for i in set(labels)}
    
    # Populate cluster_attributes dictionary
    for i, label in enumerate(labels):
        cluster_attributes[label].append(data[i])
    
    # Calculate statistics or metrics for each cluster
    for cluster_label, cluster_data in cluster_attributes.items():
        cluster_data = np.array(cluster_data)
        cluster_mean = np.mean(cluster_data, axis=0)
        cluster_std = np.std(cluster_data, axis=0)
        cluster_max = np.max(cluster_data, axis=0)
        cluster_min = np.min(cluster_data, axis=0)
        print(f"Cluster {cluster_label}:")
        for i in range(len(feature_names)):
            print(f"  {feature_names[i]}:")
            print(f"    Mean: {cluster_mean[i]}")
            print(f"    Standard Deviation: {cluster_std[i]}")
            print(f"    Max: {cluster_max[i]}")
            print(f"    Min: {cluster_min[i]}")
        print("")
        
cluster_analysis(data,labels,feature_names)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Use elbow method to figure out optimal number of clusters
def elbow_method(data, max_clusters=10):
    wcss = []
    silhouette_scores = []
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
        if i > 1:
            silhouette_scores.append(silhouette_score(data, kmeans.labels_))
        else:
            silhouette_scores.append(0)  
    return wcss, silhouette_scores

def plot_elbow(wcss, silhouette_scores, max_clusters=10):
    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('Number of Clusters')
    ax1.set_ylabel('WCSS', color=color)
    ax1.plot(range(1, max_clusters + 1), wcss, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  
    plt.title('Elbow Method')
    plt.show()

data = [argentina, australia, austria, belarus, belgium, bolivia, brazil, bulgaria, canada, chile, costa_rica, czech_republic, denmark, dominican_republic, ecuador, egypt, el_salvador, estonia, finland, france, germany,greece,guatemala,honduras,hong_kong,hungary,iceland,india,indonesia,ireland,israel,italy,japan,kazakhstan,latvia,lithuania, malaysia, mexico, morocco, the_netherlands, new_zealand, nicaragua, nigeria, norway, pakistan, panama, paraguay, peru, philippines, poland, portugal, romania, saudi_arabia, singapore, slovakia, south_africa, south_korea, spain, sweden, switzerland, taiwan, thailand, turkey, uae, usa, ukraine, uruguay,venezuela, vietnam, billboard]

wcss, silhouette_scores = elbow_method(data)
plot_elbow(wcss, silhouette_scores)
```
